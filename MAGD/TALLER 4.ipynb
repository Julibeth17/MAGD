```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,3,6,7,10])

plt.scatter(x, y)
plt.show()
```

```python

X=np.vstack([np.ones_like(x), x]).T
print(X)
```

```python
p1=np.dot((np.transpose(X)),X)
p2=np.linalg.inv(p1)
p3=np.dot(p2,np.transpose(X))
Beta=np.dot(p3,y)
print(Beta)
Inercepto=Beta[0]
Coeficiente=Beta[1]
y_predct=np.dot(X,Beta)
print(y_predct)

plt.scatter(x,y, label="Datos")
plt.plot(x,y_predct, color="red", label="Regresi√≥n Lineal")
plt.legend()
plt.show()
```
